```python
import pandas as pd
```

```python
users_df = pd.read_csv('users.csv',sep=r'\s*(?:\t|,)\s*',engine='python')
experiments_df = pd.read_csv('user_experiments.csv',sep=r'\s*(?:\t|,)\s*',engine='python')
compounds_df = pd.read_csv('compounds.csv',sep=r'\s*(?:\t|,)\s*',engine='python')
```

```python
users_df.head()
```

```python
total_experiments = experiments_df.groupby('user_id')["experiment_id"].count().reset_index(name="total_experiments")
```

```python
total_experiments
```

```python
average_experiments = experiments_df.groupby('user_id')["experiment_run_time"].mean().reset_index(name='average_experiments_amount')
```

```python
average_experiments
```

```python
common_compound = experiments_df.groupby('user_id')['experiment_compound_ids'].apply(lambda x: x.str.split(';').explode().mode()[0]).reset_index(name='common_compound')
```
